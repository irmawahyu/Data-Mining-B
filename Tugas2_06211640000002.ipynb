## Imputation Missing Value

```python
np.sum(data.isnull()) #number of NA for each variable
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
#imputasi missing value data numerik
data['Budget'] = data['Budget'].fillna((data['Budget'].mean())) 
data['Screens'] = data['Screens'].fillna((data['Screens'].mean()))
data['Aggregate Followers'] = data['Aggregate Followers'].fillna((data['Aggregate Followers'].mean()))
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(data.mean()))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk Sex, Steroid, Anaroxia, Spleen Pabable, Spiders tanpa membuat kolom baru
data['Year'] = LabelEncoder().fit_transform(data['Year'])
data['Movie'] = LabelEncoder().fit_transform(data['Movie'])
data['Genre'] = LabelEncoder().fit_transform(data['Genre'])
data.head()
```
